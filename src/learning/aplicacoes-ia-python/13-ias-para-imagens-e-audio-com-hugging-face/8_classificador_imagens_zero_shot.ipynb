```python
from transformers import pipeline 
modelo = "openai/clip-vit-large-patch14"
classificador = pipeline("zero-shot-image-classification", model=modelo)
```

```python
def mostrar_imagem(imagem):

    imagem = imagem.copy()	
    imagem.thumbnail((250, 250))
    display(imagem)
```

```python

from pathlib import Path 
from PIL import Image

animais = [Image.open(arquivo) for arquivo in Path("imagens/animals").iterdir()] 
animais
```

```python
classes = ["wolf", "tiger", "dog", "cat", "puma", "turtle", "bird"]

for imagem in animais:
    predicao = classificador(imagem, candidate_labels=classes)
    print("Predição")
    for pred in predicao:
        print(f'{pred["label"]}: {100 * pred["score"]:.2f}%')
    mostrar_imagem(imagem)
```

```python
ambientes = [Image.open(arquivo) for arquivo in Path("imagens/casas").iterdir()]
ambientes
```

```python
classes = [
    "common bathroom",
    'luxury bathroom',
    'common living room',
    'luxury living room',
    'common kitchen',
    'luxury kitchen',
    'outdoor room',
    'common building',
    'tall building',
    'old building',
    'small building',
    'museum',
]

for imagem in ambientes: 	
	predicao = classificador(imagem, candidate_labels=classes) 
	print("Predição")
	for pred in predicao:
		print(f'{pred["label"]}: {100 * pred["score"]:.2f}%')
  	
	mostrar_imagem(imagem)
```
