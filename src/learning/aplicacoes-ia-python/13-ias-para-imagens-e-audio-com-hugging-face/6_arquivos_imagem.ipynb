```python
from datasets import load_dataset
dados = load_dataset('cats_vs_dogs', split="train", streaming=True)
```

```python
for linha in dados.take(5):
    print(linha)
```

```python
imagem = list(dados.take(5))[3]["image"]
imagem
```

```python
imagem.format
```

```python
imagem.mode
```

```python
imagem.width, imagem.height
```

```python
import numpy as np 
dados_imagem = np.array(imagem)
dados_imagem
```

```python
dados_imagem.shape
```

```python
dados_r = dados_imagem[:, :,0]
dados_g = dados_imagem[:, :,1]
dados_b = dados_imagem[:, :,2]
```

```python
import matplotlib.pyplot as plt 

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2)

ax1.imshow(imagem)
ax2.imshow(dados_r, cmap='Reds', vmin=0, vmax=255)
ax3.imshow(dados_g, cmap='Greens', vmin=0, vmax=255)
ax4.imshow(dados_b, cmap='Blues', vmin=0, vmax=255)

plt.show()
```
