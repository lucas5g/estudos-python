```python
# No se preocupe em entender este cdigo!
#  apenas para o diagrama abaixo
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# Cria o dataset
dataset = np.random.randn(25)
# Cria outro rugplot
sns.rugplot(dataset);
# Configure o eixo dos x para o grfico
x_min = dataset.min() - 2
x_max = dataset.max() + 2
# 100 pontos igualmente espaados de x_min para x_max
x_axis = np.linspace(x_min,x_max,100)
# Configure a largura de banda. Para obter informaes sobre isso:

url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'
bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2
# Crie uma lista de kernel vazia
kernel_list = []
# Traa cada funo de base
for data_point in dataset:
	# Crie um kernel para cada ponto e acrescente  lista
	kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
	kernel_list.append(kernel)
	# Ajusta a escala para plotar
	kernel = kernel / kernel.max()
	kernel = kernel * .4
	plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)
plt.ylim(0,1)
```

```python
# Para obter o grfico do kde podemos somar essas funes de base.
# Traa a soma da funo de base
sum_of_kde = np.sum(kernel_list,axis=0)
# Plota a figura
fig = plt.plot(x_axis,sum_of_kde,color='indianred')
# Adiciona o rugplot inicial
sns.rugplot(dataset,c = 'indianred')
# Livrar-se das marcas de "y-tick"
plt.yticks([])
# Define o ttulo
plt.suptitle("Sum of the Basis Functions")
```
